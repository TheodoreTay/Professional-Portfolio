```python
# Dropped the following from x - SaleCondition,Id,SalePrice
y = house_three['SaleCondition']
no_price = house_three.drop(labels=['SaleCondition','Id','SalePrice','GarageYrBlt'],axis=1).copy()
```

```python
no_price_dum = pd.get_dummies(no_price,drop_first=True)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
lr = LogisticRegression()
lr.fit(X_final, y)
```

```python
gs_params = {
    'penalty':['l1','l2'],
    'solver':['liblinear'],
    'C':np.logspace(-5,0,100)
}
```

```python
from sklearn.model_selection import GridSearchCV
lr_gridsearch = GridSearchCV(lr, gs_params, cv=5, verbose=1)
```

```python
lr_gridsearch.fit(X_final, y)
```

```python
print ('best score',lr_gridsearch.best_score_)
print ('best param', lr_gridsearch.best_params_)
```

```python
best_lr = lr_gridsearch.best_estimator_
coef_df = pd.DataFrame({
        'coef':best_lr.coef_[0],
        'feature':no_price_dum.columns
    })
```

```python
coef_df['abs_coef'] = np.abs(coef_df.coef)
coef_df.sort_values('abs_coef', ascending=False, inplace=True)
```
